## histogram

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
x = [1, 3, 5, 1, 2, 4, 4, 2, 5, 4, 3, 1, 2]
n_bins = 5
plt.hist(x, bins=n_bins)
plt.show()
```

```python
np.random.seed(31415)
n_points = 10000
n_bins = 15
x = np.random.randn(n_points)
plt.hist(x, bins=n_bins)
plt.show()
```

```python
y = np.random.randn(n_points)
plt.hist2d(x, y, bins=50)
plt.show()
```

```python
plt.hist(x, 20, density=True,
         histtype='stepfilled',
         facecolor='g', alpha=0.5)
plt.show()
```

```python
plt.hist(x, 20, density=True,
         histtype='step')
plt.show()
```

## contours


```python
x = np.arange(-3, 3, 0.005)
y = np.arange(-3, 3, 0.005)
X, Y = np.meshgrid(x, y)
Z = (X**2 + Y**2)
out = plt.contour(X, Y, Z)
plt.clabel(out, inline=True,
           fontsize=10)
plt.show()
```

```python
out = plt.contour(X, Y, Z)
plt.clabel(out, inline=True,
           fontsize=10)
plt.colorbar(out)
plt.show()
```

```python
out = plt.contour(X, Y, Z,
                  colors='g')
plt.clabel(out, inline=True,
           fontsize=10)
plt.show()
```

```python
plt.contourf(X, Y, Z,
             hatches=['-', '/', '\\', '//'],
             cmap='cool',
             alpha=0.75)
#err plt.colorbar(X)
plt.show()
```

vector

```python
Y, X = np.mgrid[-5:5:200j, -5:5:300j] 
U = X**2 + Y**2
V=X+ Y
plt.streamplot(X, Y, U, V)
plt.show()
```

```python
plt.streamplot(X, Y, U, V, density=[0.5, 0.75])
plt.show()
```

```python
plt.streamplot(X, Y, U, V, color=V,
           linewidth=1, cmap='cool')
plt.show()
```

```python
plt.streamplot(X, Y, U, V,density=0.6,color='k', linewidth=X)
plt.show()
```

```python
plt.quiver(X, Y, U, V)
plt.show()
```

```python
X = np.arange(-5, 5, 0.5)
Y = np.arange(-10, 10, 1)
U, V = np.meshgrid(X, Y)
plt.quiver(X, Y, U, V)
plt.show()
```

## hyperbola

```python
# hyperbola
Y, X = np.mgrid[-5:5:200j, -5:5:300j] 
U = X**2 - Y**2
V=X+ Y
plt.streamplot(X, Y, U, V)
plt.show()
```

```python
plt.streamplot(X, Y, U, V, density=[0.5, 0.75])
plt.show()
```

```python
plt.streamplot(X, Y, U, V, color=V,
           linewidth=1, cmap='cool')
plt.show()
```

```python
plt.streamplot(X, Y, U, V,
               density=0.6,
               color='k',
               linewidth=X)
plt.show()
```

## mesh grid vector

```python
X = np.arange(-5, 5, 0.5)
Y = np.arange(-10, 10, 1)
U, V = np.meshgrid(X, Y)
plt.quiver(X, Y, U, V)
plt.show()
```

```python

X = np.arange(-5, 5, 0.5)
Y = np.arange(-10, 10, 1)
#U, V = np.meshgrid(X, Y)
U = X**2 - Y**2
V=X+ Y
plt.quiver(X, Y, U, V)
plt.show()
```

```python
X = np.arange(-5, 5, 0.5)
Y = np.arange(-10, 10, 1)
U, V = np.meshgrid(X**-Y**2,X+Y)
plt.quiver(X, Y, U, V)
plt.show()
```
